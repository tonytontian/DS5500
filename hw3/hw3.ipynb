# Question 3
The district-level performance metrics from EDFacts may be useful in your decision.

However, to protect student privacy, the data in these datasets has been heavily “blurred” to prevent students
from being identified. Therefore, most of the numeric metrics are presented as ranges in string format. In
addition, censored and missing data must be imputed.

Write and explain a function for processing a single column of “blurred” metrics into usable numeric values.
Use it to process and then visualize the distribution of a performance metric of your choice.


There are 5 different kind of data in the blurred field. 

PS --> remove row because no valid data.

GE --> greater or equals, add 2, up to 100

GT --> greater than, add 5, up to 100

LE --> less or equals, substruct 2, down to 0

LT --> less than, substruct 5, down to 0

```python
data_math = pd.read_csv("./data/math-achievement-lea-sy2015-16.csv")
data_math_pctprof = data_math[["ALL_MTH00PCTPROF_1516"]].query('ALL_MTH00PCTPROF_1516 != "PS"')
```

```python
def unblur(blurred_data):
    if('-' in blurred_data):
        temp = blurred_data.split('-');
        return (int(temp[0]) + int(temp[1])) / 2
    elif('GE' in blurred_data):
        return min(int(blurred_data.lstrip('GE')) + 2, 100)
    elif('GT' in blurred_data):
        return min(int(blurred_data.lstrip('GT')) + 5, 100)
    elif('LE' in blurred_data):
        return max(int(blurred_data.lstrip('LE')) - 2, 0)
    elif('LT' in blurred_data):
        return max(int(blurred_data.lstrip('LT')) - 5, 0)
    else:
        return int(blurred_data)
```

```python
data_math_pctprof["unblurred_pctprof"] = data_math_pctprof.apply(lambda x: unblur(x["ALL_MTH00PCTPROF_1516"]), axis=1)
```

```python
data_math_pctprof["unblurred_pctprof"].hist(bins=20)
```
